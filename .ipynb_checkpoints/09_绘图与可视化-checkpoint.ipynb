## 一、Matplotlib的基本用法

### （一）导入模块

使用import导入模块matplotlib.pyplot，并简写成plt；使用import导入模块numpy，并简写成np

```python
#coding:utf-8

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.sans-serif']=['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
```

然后创建两组数据，使用np.linspace定义x：范围是(-3,3)，个数是50，将产生一组（-3，3）内均匀分布的50个数；(x,y1)表示曲线1，(x,y2)表示曲线2。

```python
x = np.linspace(-3, 3, 50)
y_1 = 2*x + 1
y_2 = x**2
```

### （一）定义窗口并画图

在画图前使用plt.figure()定义一个图像窗口：编号为3；大小为(8, 5)；这两项参数可缺省。其中，num参数决定了程序运行后弹出的图像窗口名字，但在klab平台下不会显示。接着，我们使用plt.plot画出(x ,y2)曲线；使用plt.plot画(x ,y1)曲线，曲线的颜色属性(color)为红色；曲线的宽度(linewidth)为1.0；曲线的类型(linestyle)为虚线，除了虚线外，大家还可使用以下线性：'-'、'--'、'-.'、':' 。接着，我们使用plt.show()显示图像。

```python
plt.figure(num=3, figsize=(8,5))  # 定义图像窗口，编号为3，大小为(8, 5)
plt.plot(x, y_2) # 画图
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='-.')  # color为颜色，linestyle可以为'-','--','-.',':'
plt.show()  # 显示图像
```

### （二）定义坐标轴名称及范围

使用plt.xlim设置x坐标轴范围：(-1, 2)； 使用plt.ylim设置y坐标轴范围：(-2, 3)； 使用plt.xlabel设置x坐标轴名称：’I am x’； 使用plt.ylabel设置y坐标轴名称：’I am y’；

```python
plt.figure(num=3, figsize=(8,5))

# 画图
plt.plot(x, y_2)
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='-.')

# 设置坐标轴范围
plt.xlim(-1, 2)
plt.ylim(-2, 3)

# 设置坐标轴标签
plt.xlabel('I am x')
plt.ylabel('I am y')

plt.show()
```

### （三）定义坐标轴刻度及名称

有时候，我们的坐标轴刻度可能并不是一连串的数字，而是一些文字，或者我们想要调整坐标轴的刻度的稀疏，这时，就需要使用plt.xticks()或者plt.yticks()来进行调整：首先，使用np.linspace定义新刻度范围以及个数：范围是(-1,2);个数是5。使用plt.xticks设置x轴刻度：范围是(-1,2)，个数是5。使用plt.yticks设置y轴刻度以及名称：刻度为[-2, -1.8, -1, 1.22, 3]；对应刻度的名称为[‘really bad’,’bad’,’normal’,’good’, ‘really good’]。使用plt.show()显示图像。

```python
plt.figure(num=3, figsize=(8,5))

plt.plot(x, y_2)
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='-.')

# 设置坐标轴范围
plt.xlim(-1,2)
plt.ylim(-2,3)

# 设置坐标轴label
plt.xlabel('I am x')
plt.ylabel('I am y')

# 定义新刻度
x_new_ticks = np.linspace(-1,2,5)
y_new_ticks, y_ticks_name = [-2, -1.8, -1, 1.22, 3], ['really bad', 'bad', 'normal', 'good', 'really good']

print(x_new_ticks)
print(y_new_ticks)

# 重新定义刻度
plt.xticks(x_new_ticks)
plt.yticks(y_new_ticks, y_ticks_name)

plt.show()
```

### （四）设置图像边框颜色

观察以上图像，我们的图像坐标轴总是由上下左右四条线组成，我们也可以对它们进行修改：首先，使用plt.gca()获取当前坐标轴信息。使用.spines设置边框：右侧边框；使用.set_color设置边框颜色：默认白色； 使用.spines设置边框：上边框；使用.set_color设置边框颜色：默认白色。

```python
plt.figure(num=3, figsize=(8,5))

plt.plot(x, y_2)
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='-.')

# 设置坐标轴范围
plt.xlim(-1,2)
plt.ylim(-2,3)

# 设置坐标轴label
plt.xlabel('I am x')
plt.ylabel('I am y')

# 定义新刻度
x_new_ticks = np.linspace(-1,2,5)
y_new_ticks, y_ticks_name = [-2, -1.8, -1, 1.22, 3], ['really bad', 'bad', 'normal', 'good', 'really good']

print(x_new_ticks)
print(y_new_ticks)

# 重新定义刻度
plt.xticks(x_new_ticks)
plt.yticks(y_new_ticks, y_ticks_name)

# 获取坐标轴信息
ax = plt.gca()
print(ax)

# 设置边框
print(ax.spines)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

### （五）调整刻度及边框位置

使用.xaxis.set_ticks_position设置x坐标刻度数字或名称的位置：bottom.（所有位置：top，bottom，both，default，none）；使用.spines设置边框：x轴；使用.set_position设置边框位置：y=0的位置；（位置所有属性：outward，axes，data）

```python
plt.figure(num=3, figsize=(8,5))

plt.plot(x, y_2)
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle='-.')

# 设置坐标轴范围
plt.xlim(-1,2)
plt.ylim(-2,3)

# 设置坐标轴label
plt.xlabel('I am x')
plt.ylabel('I am y')

# 定义新刻度
x_new_ticks = np.linspace(-1,2,5)
y_new_ticks, y_ticks_name = [-2, -1.8, -1, 1.22, 3], ['really bad', 'bad', 'normal', 'good', 'really good']

print(x_new_ticks)
print(y_new_ticks)

# 重新定义刻度
plt.xticks(x_new_ticks)
plt.yticks(y_new_ticks, y_ticks_name)

# 获取坐标轴信息
ax = plt.gca()
print(ax)

# 设置边框
print(ax.spines)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 设置x轴的刻度位置
ax.xaxis.set_ticks_position('bottom')
'''
position{'top', 'bottom', 'both', 'default', 'none'}
'both' sets the ticks to appear on both positions, but does not change the tick labels. 'default' resets the tick 
positions to the default: ticks on both positions, labels at bottom. 'none' can be used if you don't want any ticks.
'none' and 'both' affect only the ticks, not the labels.
'''

# 设置边框位置
ax.spines['bottom'].set_position(('data', 0))

'''
'outward': place the spine out from the data area by the specified number of points. (Negative values place the spine inwards.)
'axes': place the spine at the specified Axes coordinate (0 to 1).
'data': place the spine at the specified data coordinate.
'''

plt.show()
```

### 一个练习

画出直线 y = x-1, 线型为虚线，线宽为1，纵坐标范围（-2，2），横坐标范围（-1，2），横纵坐标在（0，0）坐标点相交。横坐标的 [-1,-0.5,1] 分别对应 [bad, normal, good]。

## 二、图例和标注

### （一）添加图例

首先我们设置两条线的类型等信息（蓝色实线与红色虚线)，并且通过label参数为两条线设置名称。比如直线的名称就叫做 "一次函数", 曲线的名称叫做 "二次函数"。当然，只是设置好名称并不能使我们的图例出现，要通过plt.legend()设置图例的显示。legend获取代码中的 label 的信息, plt 就能自动的为我们添加图例。

```python
x = np.linspace(-3, 3, 50)
y_1 = 2*x + 1
y_2 = x**2

plt.figure()

# 设置坐标轴范围
plt.xlim(-1, 2)
plt.ylim(-2, 3)

# 设置新的刻度
x_new_ticks = np.linspace(-1, 2, 5)
y_new_ticks, y_ticks_name = [-2, -1.8, -1, 1.22, 3], ['really bad', 'bad', 'good', 'really good']
plt.xticks(x_new_ticks)
plt.yticks(y_new_ticks, y_ticks_name)

# 画图
plt.plot(x, y_2, label='二次函数')
plt.plot(x, y_1, color='red', linewidth=1.0, linestyle=':', label='一次函数')

# 设置图例
plt.legend(loc='lower right') 
'''
loc参数调整图例位置，
'best' : 0, 'upper right' : 1, 'upper left' : 2, 'lower left' : 3, 
'lower right' : 4, 'right' : 5, 'center left' : 6, 'center right' : 7,
'lower center' : 8, 'upper center' : 9, 'center' : 10
'''

plt.show()
```

同样可以通过设置 handles 参数来选择图例中显示的内容。首先，在上面的代码 plt.plot(x, y2, label='linear line') 和 plt.plot(x, y1, label='square line') 中用变量 l1 和 l2 分别存储起来，而且需要注意的是 l1, l2,要以逗号结尾, 因为plt.plot() 返回的是一个列表。然后将 l1,l2 这样的objects以列表的形式传递给 handles。另外，label 参数可以用来单独修改之前的 label 信息, 给不同类型的线条设置图例信息。

```python
# 画图
l1, = plt.plot(x, y_2, label='二次函数')
l2, = plt.plot(x, y_1, color='red', linewidth=1.0, linestyle=':', label='一次函数')

# 设置图例
plt.legend(handles=[l1, l2,], labels=['linear line', 'square line'], loc='lower right')

plt.show()
```

### （二）Annotation 标注

当图线中某些特殊地方需要标注时，我们可以使用 annotation. matplotlib 中的 annotation。有两种方法，一种是用 plt 里面的 annotate，一种是直接用 plt 里面的 text 来写标注。

#### 1. plt.annotate()

标注出点(x0, y0)的位置信息。用plt.plot([x_0, x_0], [0, y_0], 'k--', linewidth=2.5) 画出一条垂直于x轴的虚线。其中，[x_0, x_0], [0, y_0] 表示在图中画一条从点 (x_0,0) 到 (x_0,y_0) 的直线，'k--' 表示直线的颜色为黑色(black)，线形为虚线。而 plt.scatter 函数可以在图中画点，此时我们画的点为 (x_0,y_0), 点的大小(size)为 50， 点的颜色为蓝色(blue),可简写为 b。

对(x0, y0)这个点进行标注。第一种方式就是利用函数 annotate()，其中 '2x+1={}'.format(y_0) 代表标注的内容；参数xycoords='data' 是说基于数据的值来选位置, xytext=(+30, -30) 和 textcoords='offset points' 表示对于标注位置的描述 和 xy 偏差值，即标注位置是 xy 位置向右移动 30，向下移动30, arrowprops是对图中箭头类型和箭头弧度的设置，需要用 dict 形式传入。

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)

ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

x_0 = 1
y_0 = 2*x_0 + 1

# 画出点(x_0, 0)与(x_0, y_0)之间的线段
plt.plot([x_0, x_0], [0, y_0], 'k--', linewidth=2.5)

# 画出点(x_0, y_0)
plt.scatter(x_0, y_0, s=50, color='b')

# 进行标注
plt.annotate('2x+1={}'.format(y_0), xy=(x_0, y_0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

plt.plot(x, y)

plt.show()
```

#### 2. plt.text()

第二种注释方式是通过 text() 函数，其中 -3.7,3, 是选取text的位置, r'' 为 text 的内容，其中空格需要用到转字符 \ ,fontdict 设置文本字的大小和颜色。

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)

ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

x_0 = 1
y_0 = 2*x_0 + 1

# 画出点(x_0, 0)与(x_0, y_0)之间的线段
plt.plot([x_0, x_0], [0, y_0], 'k--', linewidth=2.5)

# 画出点(x_0, y_0)
plt.scatter(x_0, y_0, s=50, color='b')

# 进行标注
plt.annotate('2x+1={}'.format(y_0), xy=(x_0, y_0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

# 标注：r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$'  其中r''、\、$ 是必须的，否则mu、sigma_i、alpha_t将无法显示成数学符号
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$', fontdict={'size': 16, 'color': 'black'})

plt.plot(x, y)

plt.show()
```

### （三）tick 能见度

当图片中的内容较多，相互遮盖时，我们可以通过设置相关内容的透明度来使图片更易于观察，也即是通过本节中的bbox参数设置来调节图像信息。

然后对被遮挡的图像调节相关透明度，本例中设置 x轴 和 y轴 的刻度数字进行透明度设置。其中label.set_fontsize(12)重新调节字体大小，bbox设置目标内容的透明度相关参数，facecolor调节 box 前景色，edgecolor 设置边框， 本处设置边框为无，alpha设置透明度. 最终结果如下:

```python
x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure()

# 在 plt 2.0.2 或更高的版本中, 设置 zorder 给 plot 在 z 轴方向排序
plt.plot(x, y, linewidth=10, zorder=1)

plt.ylim(-2, 2)

ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.7, zorder=2))

plt.show()
```

## 三、各类图形绘制

### （一）散点图 plt.scatter()

#### 1. 生成数据集

生成1024个呈标准正态分布的二维数据组 (平均数是0，方差为1) 作为一个数据集，并图像化这个数据集。每一个点的颜色值用color_num来表示。

```python
n = 1024
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
color_num = np.arctan2(x,y)  # 设置颜色数值
```

#### 2. 画图

数据集生成完毕，现在来用 plt.scatter 画出这个点集，输入x和y作为location，size=75，颜色为color_num，color map用默认值，透明度alpha 为 50%。 x轴显示范围定位(-1.5，1.5)，并向xtick()函数传入空集()来隐藏x坐标轴，y轴同理。

```python
plt.figure()

plt.scatter(x, y, s=75, c=color_num, alpha=0.5)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.xticks(())
plt.yticks(())

plt.show()
```

### （二）柱状图 plt.bar()

#### 1. 生成数据集

首先生成画图数据，向上向下分别生成2组数据，x为0到11的整数 ，y是相应的均匀分布的随机数据。

```python
n = 12
x = np.arange(n)
y = (1 - x / float(n)) * np.random.uniform(0.5, 1.0, n)
```

#### 2. 画图

使用的函数是plt.bar，参数为x和y，x代表横坐标，即柱形的位置，y代表纵坐标，即柱形的高度。

```python
plt.figure()

plt.bar(x, y)
plt.bar(x, -y)

plt.xlim(-0.5, n)
plt.ylim(-1.25, 1.25)

plt.xticks(())
plt.yticks(())

plt.show()
```

#### 3. 修改颜色和数据标签

如果想要改变柱状图的颜色，并且希望每个柱形上方能够显示该项数值该怎么做呢？我们可以用 plt.bar 函数中的facecolor参数设置柱状图主体颜色，用edgecolor参数设置边框颜色；而函数 plt.text 可以帮助我们在柱体上方（下方）加上数值：用%.2f保留两位小数，用ha='center'设置横向居中对齐，用va='bottom'设置纵向底部（顶部）对齐。

```python
plt.figure()

plt.bar(x, y, facecolor='#FFDAAA', edgecolor='white')
plt.bar(x, -y, facecolor='#6693CC', edgecolor='white')

for i, j in zip(x, y):
    print((i, j))
    plt.text(i, j, '%.2f' % j, ha='center', va='bottom')

for i, j in zip(x,y):
    plt.text(i, -j, '%.2f' % j, ha='center', va='top')

plt.show()
```

### （三）等高线图 plt.contourf()

#### 1. 数据生成

数据集即三维点 (x,y) 和对应的高度值，共有256个点。高度值使用一个 height function f(x,y) 生成。 x, y 分别是在区间 [-3,3] 中均匀分布的256个值，并用meshgrid在二维平面中将每一个x和每一个y分别对应起来，编织成栅格。

```python
def f(x,y):
    # the height function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y)
print(X)
print(Y)
```

#### 2. 画图

接下来进行颜色填充。使用函数plt.contourf把颜色加进去，位置参数分别为：X, Y, f(X,Y)。8代表等高线的密集程度，这里被分为10个部分。如果是0，则图像被一分为二。透明度为0.75，并将 f(X,Y) 的值对应到colormap的RdBu组中寻找对应颜色。大家可能并不能直观理解 colormap ，它可以将颜色和数字进行映射。如果暂时不能理解的话也没有关系，我们可以将其想象成matplotlib为我们提供的配色方案，大家可以查看此[链接](https://matplotlib.org/examples/color/colormaps_reference.html)选择自己喜欢的配色方案应用在自己的图上。

```python
plt.figure()

# print(f(x,y))
plt.contourf(X, Y, f(X, Y), 8, alpha=0.75, cmap=plt.cm.RdBu)

plt.show()
```

接下来进行等高线绘制。使用plt.contour函数划线。位置参数为：X, Y, f(X,Y)。颜色选黑色，线条宽度选0.5。现在的结果如下图所示，只有颜色和线条，还没有数值Label：

```python
plt.figure()

# print(f(x,y))
plt.contourf(X, Y, f(X, Y), 8, alpha=0.75, cmap=plt.cm.RdBu)

# 绘制等高线
plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=0.5)  

plt.show()
```

#### 3. 添加高度数字

最后通过 plt.clabel() 在等高线上加入高度数值，即加入Label，其中参数 inline 控制是否将 Label 画在线里面，fontsize 设置字体大小为10。并将坐标轴隐藏

### （四）随机矩阵画图 plt.imshow()

#### 1. 数据生成

首先生成一个 3x3 的 2D-array ，也就是三行三列的格子，array 中的每个值经过colormap与一个颜色对应并填充在格子中。

```python
a = np.random.randn(3,3)
print(a)
```

#### 2. 画图

之前选cmap的参数时用的是：cmap=plt.cmap.RdBu，而现在，我们可以直接用单引号传入参数。 origin='lower'代表的就是选择的原点的位置。而 interpolation 表示画图方式，从该[链接](https://matplotlib.org/examples/images_contours_and_fields/interpolation_methods.html)可看到matplotlib官网上对于内插法的不同方法的描述。这里我们使用的是内插法中的 Nearest-neighbor 的方法，其他的方式也都可以随意取选。

```python
plt.figure()

plt.imshow(a, interpolation='nearest', cmap='RdBu', origin='lower')

plt.show()
```

#### 3. 增加colorbar

下面添加一个colorbar ，它可以为我们显示不同颜色的区块所对应的具体数值。其中shrink参数可以用来调整 colorbar 的长度，这里我们使colorbar的长度变短为原来的92%，这样2D图像就创建完毕了。

```python
plt.figure()

plt.imshow(a, interpolation='nearest', cmap='RdBu', origin='lower')

# 添加颜色值
plt.colorbar(shrink=0.92)

plt.xticks(())
plt.yticks(())

plt.show()
```

#### 1.plt.Subplot()

**均匀图中图**

使用plt.figure创建一个图像窗口.使用plt.subplot来创建小图。plt.subplot(2,2,1)表示将整个图像窗口分为2行2列, 当前位置为1. 使用plt.plot([0,1],[0,1])在第1个位置创建一个小图。plt.subplot(2,2,2)表示将整个图像窗口分为2行2列, 当前位置为2。使用plt.plot([0,1],[0,2])在第2个位置创建一个小图。plt.subplot(2,2,3)表示将整个图像窗口分为2行2列,当前位置为3。 plt.subplot(2,2,3)可以简写成plt.subplot(223), matplotlib同样可以识别。使用plt.plot([0,1],[0,3])在第3个位置创建一个小图。plt.subplot(224)表示将整个图像窗口分为2行2列, 当前位置为4。使用plt.plot([0,1],[0,4])在第4个位置创建一个小图。

```python
plt.figure()

plt.subplot(2,2,1)
plt.plot([0,1],[0,1])

plt.subplot(2,2,2)
plt.plot([0,1],[0,2])

plt.subplot(223)
plt.plot([0,1],[0,3])

plt.subplot(224)
plt.plot([0,1],[0,4])

plt.show()  # 展示
```

**不均匀图中图**

#### 2. plt.subplots()

首先，使用plt.subplots建立一个2行2列的图像窗口，sharex=True表示共享x轴坐标, sharey=True表示共享y轴坐标. ((ax11, ax12), (ax21, ax22))表示第1行从左至右依次放ax11和ax12, 第2行从左至右依次放ax21和ax22.接着使用ax11.scatter创建一个散点图。

```python
f, ((ax11, ax12), (ax21, ax22)) = plt.subplots(2, 2)

# 在ax11上绘制散点图
x = np.random.normal(0,1,256)
y = np.random.normal(0,1,256)
color_num = np.arctan2(np.exp(x),np.exp(y))  # 设置颜色值
ax11.scatter(x, y, c=color_num, alpha=0.5)

# 在ax22上绘制柱状图
x = np.arange(10)
y = x**(1/2)
ax22.bar(x, y)

plt.show() # 要放在最后，不然前面的图不能全部显示
```

#### 3. plt.subplot2grid()

```python
x = np.arange(9)
y = x**2

plt.figure(figsize=(12,10))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax1.set_title('ax1')
ax1.plot(x,y)

ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax2.set_title('ax2')
ax2.plot(x,y)

ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax3.set_title('ax3')
ax3.plot(x,y)

ax4 = plt.subplot2grid((3,3), (2,0))
ax4.set_title('ax4')
ax4.plot(x,y)

ax5 = plt.subplot2grid((3,3), (2,1))
ax5.set_title('ax5')
ax5.plot(x,y)

plt.show()
```

#### 4. matplotlib.gridspec模块

gridspec 同样能帮助我们画出前文的图，它允许我们使用索引的方式指定小图的大小和位置。

使用plt.figure()创建一个图像窗口, 使用gridspec.GridSpec将整个图像窗口分成3行3列.使用plt.subplot来作图, gs[0, :]表示这个图占第0行和所有列, gs[1, :2]表示这个图占第1行和第2列前的所有列, gs[1:, 2]表示这个图占第1行后的所有行和第2列, gs[-1, 0]表示这个图占倒数第1行和第0列, gs[-1, -2]表示这个图占倒数第1行和倒数第2列.

```python
# 导入模块
import matplotlib.gridspec as gridspec

# 生成数据
x = np.arange(9)
y = x**(1/2)

plt.figure(figsize=(10, 10))

# 将整个图像窗口分成3行3列
gs = gridspec.GridSpec(3, 3)

ax1 = plt.subplot(gs[0, :])
ax1.set_title('ax1')
ax1.plot(x,y)

ax2 = plt.subplot(gs[1, :2])
ax2.set_title('ax2')
ax2.plot(x,y)

ax3 = plt.subplot(gs[1:, 2])
ax3.set_title('ax3')
ax3.plot(x,y)

ax4 = plt.subplot(gs[-1, 0])
ax4.set_title('ax4')
ax4.plot(x,y)

ax5 = plt.subplot(gs[-1, -2])
ax5.set_title('ax5')
ax5.plot(x,y)

plt.show()
```

### （二）图中图

```python
# 生成数据
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

## 绘制大图
# 确定大图左下角的位置以及宽高  
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
'''注意，4个值都是占整个figure窗口的百分比。在这里，假设figure的大小是10x10，
那么大图就被包含在由(1, 1)开始，宽8，高8的坐标系内。'''
# 大图坐标系添加到 figure 中，颜色为 lightskyblue，取名为 title
fig = plt.figure()
ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, 'lightskyblue')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('外层大图')

## 绘制小图
# 确认小图位置与宽高
left, bottom, width, height = 0.2, 0.6, 0.2, 0.2
# 画小图
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(x, y, 'lightslategrey')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('内层小图_1')

## 另一种方法画内层小图
left, bottom, width, height = 0.6, 0.2, 0.2, 0.2
plt.axes([left, bottom, width, height])
plt.plot(x, y, 'lightblue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('内层小图_2')

plt.show()
```

## 五、3D作图

### 3D作图

```python
# 导入模块
from mpl_toolkits.mplot3d import Axes3D

# 编制平面网络
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
x, y = np.meshgrid(x, y)  # 生成平面网络
# print(x)
# print(y)

# 计算高度
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

# 创建3D图像窗口
fig = plt.figure()
ax = Axes3D(fig)

# 生成三维图
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))

# 将三维图投影到需要xy平面
ax.contourf(x, y, z, zdir='z', offset=-1, cmap=plt.get_cmap('rainbow'))

plt.show()
```

### 一个练习

画出3D数据 z = (x + y)^2， 并利用等高线图对其进行投影

```python
# 编制平面网络
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
x, y = np.meshgrid(x, y)  # 生成平面网络

# 计算高度
z = (x + y)**2

# 绘图
fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))

# 投影
ax.contourf(x, y, z, zdir='x', offset=-4, cmap=plt.get_cmap('rainbow'))

plt.show()
```

```python
raw_data = {'水果': ['香蕉', '苹果', '草莓', '香蕉', '苹果', '草莓', '香蕉', '苹果', '草莓'],
            '价格': [2,4,10,5,3,6,2,4,10],
            '数量': [5,3,6,2,4,10,5,3,6],}
df = pd.DataFrame(raw_data)
df

pos = list(range(len(df['水果'].unique()))) 
width = 0.25

plt.figure()

# 画价格柱状图
x1 = [p - width/2 for p in pos]
y1 = df.groupby('水果').mean()['价格']
plt.bar(x1, y1, width, color='#FFCCCC') 

for a, b in zip(x1, y1):
    plt.text(a, b+0.5, '%.2f' % b, ha='center', va='bottom', fontsize=10)

# 画数量柱状图
x2 = [p + width/2 for p in pos]
y2 = df.groupby('水果').sum()['数量']
plt.bar(x2, y2, width, color='#6699CC') 

for a, b in zip(x2, y2):
    plt.text(a, b+0.5, '%.2f' % b, ha='center', va='bottom', fontsize=10)

# 其他调整
plt.ylim(0,25)
plt.xticks(np.linspace(0, 2, 3),df['水果'])
plt.legend(['价格', '数量'], loc='upper left')

plt.show()
```
